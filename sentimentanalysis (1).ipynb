```python
import re

def clean_text(text):
    # Remove hashtags, mentions, URLs, and special characters
    text = re.sub(r"#\w+", "", text)  # Remove hashtags
    text = re.sub(r"@\w+", "", text)  # Remove mentions
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove non-alphabetical characters
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    return text

# Apply cleaning to the 'Text' column
data_cleaned["Cleaned_Text"] = data_cleaned["Text"].apply(clean_text)
```

```python
# Step 1: Normalize the Sentiment column to remove extra spaces and ensure consistent casing
data_cleaned['Sentiment'] = data_cleaned['Sentiment'].str.strip().str.capitalize()

# Step 2: Define the sentiment mapping dictionary
sentiment_mapping = {'Negative': 0, 'Neutral': 1, 'Positive': 2, 'Happy': 3}

# Step 3: Map Sentiment to numerical values
data_cleaned['Sentiment_Encoded'] = data_cleaned['Sentiment'].map(sentiment_mapping)

# Step 4: Check for any unmapped or NaN values
unmapped_values = data_cleaned[data_cleaned['Sentiment_Encoded'].isna()]
if not unmapped_values.empty:
    print("Unmapped Sentiment values:")
    print(unmapped_values['Sentiment'].unique())
```

```python
import pandas as pd
import re
import pickle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample data
data_cleaned = pd.DataFrame({
    'Sentiment': ['Positive', 'Negative', 'Neutral', 'Happy', 'Neutral'],
    'Cleaned_Text': ['I love it!', 'I hate it!', 'It is okay.', 'I feel great!', 'It is fine.']
})

# Step 1: Normalize the Sentiment column to remove extra spaces and ensure consistent casing
data_cleaned['Sentiment'] = data_cleaned['Sentiment'].str.strip().str.capitalize()

# Step 2: Define the sentiment mapping dictionary
sentiment_mapping = {'Negative': 0, 'Neutral': 1, 'Positive': 2, 'Happy': 3}

# Step 3: Map Sentiment to numerical values
data_cleaned['Sentiment_Encoded'] = data_cleaned['Sentiment'].map(sentiment_mapping)

# Step 4: Handle any unmapped values (if any)
data_cleaned['Sentiment_Encoded'] = data_cleaned['Sentiment_Encoded'].fillna(-1)  # Fill unmapped values with -1

# Text Vectorization using TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the cleaned text data
X_tfidf = tfidf_vectorizer.fit_transform(data_cleaned['Cleaned_Text'])
y = data_cleaned['Sentiment_Encoded']  # Target variable

# Split the data into training and testing sets
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Train the Logistic Regression model with class weights
logreg_model = LogisticRegression(class_weight='balanced', max_iter=1000)  # Increased max_iter and balanced class weights
logreg_model.fit(X_train_tfidf, y_train)

# Evaluate the Logistic Regression model
y_pred_logreg = logreg_model.predict(X_test_tfidf)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Logistic Regression Accuracy: {accuracy_logreg}")
print(classification_report(y_test, y_pred_logreg))

# Test prediction function with sample inputs
def preprocess_text(text):
    """
    Preprocesses the input text: cleans it by removing non-alphabetic characters and converting to lowercase.
    """
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I)
    text = text.lower()
    return text

def predict_sentiment(text, model_name='logreg_model'):
    """
    Predict sentiment using a selected model.
    """
    # Load models and vectorizer
    loaded_logreg_model = pickle.load(open('logreg_model.sav', 'rb'))
    loaded_tfidf_vectorizer = pickle.load(open('tfidf_vectorizer.sav', 'rb'))

    # Preprocess and vectorize text
    cleaned_text = preprocess_text(text)
    vectorized_text = loaded_tfidf_vectorizer.transform([cleaned_text])

    # Select model
    if model_name == 'logreg_model':
        model = loaded_logreg_model
    else:
        raise ValueError("Invalid model name.")

    # Make prediction
    prediction = model.predict(vectorized_text)
    return prediction[0]

# Test with a few examples
test_texts = ["I love this product!", "I hate this product!", "It's just okay.", "I'm feeling great!"]

for text in test_texts:
    predicted_sentiment = predict_sentiment(text, model_name='logreg_model')
    print(f"Predicted sentiment for '{text}': {predicted_sentiment}")
```
